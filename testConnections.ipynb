```python
# requirements: duckdb, huggingface_hub
import duckdb
from huggingface_hub import hf_hub_url

con = duckdb.connect()
con.execute("INSTALL httpfs; LOAD httpfs;")

# Build the final CDN URL (no download, just resolve)
url = hf_hub_url(
    repo_id="danielmlvz/water-dashboard",
    filename="consumo19/part-0.parquet",
    repo_type="dataset",
    revision="main",  # or a tag/commit
)

df = con.execute("SELECT * FROM read_parquet($url)", {"url": url}).df()
df.head()
```
